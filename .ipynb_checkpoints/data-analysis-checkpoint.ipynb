```python
def alpha3code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
            CODE.append(code)
        except:
            CODE.append('None')
    return CODE
```

```python
data2021_copy = data2021.copy()
data2021_copy['CODE']=alpha3code(data2021_copy['Country'])
```

```python
df_corr = data2021.corr()

x = list(df_corr.columns)
y = list(df_corr.index)
z = np.array(df_corr)

fig = ff.create_annotated_heatmap(
    z,
    x = x,
    y = y ,
    annotation_text = np.around(z, decimals=2),
    hoverinfo='z',
    colorscale='sunset'
    )

fig.update_layout(
    autosize=False,
    width=700,
    height=600,
   )
fig.show()
```
